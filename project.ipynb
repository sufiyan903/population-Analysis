```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Assuming you have a DataFrame named 'your_dataset'
# If you haven't loaded the dataset yet, you can do it like this:
# your_dataset = pd.read_csv("your_dataset.csv")

# Convert "Population (2020)" column to numeric
population["Population (2020)"] = pd.to_numeric(population["Population (2020)"], errors='coerce')

# Drop rows with NaN values in "Population (2020)" column
population.dropna(subset=["Population (2020)"], inplace=True)

x = population["Population (2020)"]
y = population["Density (P/Km²)"]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, slope * x + intercept, color='red')  # Adding a regression line
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.xlabel("Population (2020)")
plt.ylabel("Density (P/Km²)")
plt.title(f'Linear Regression: Population vs Density')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Assuming you have a DataFrame named 'population'
# If you haven't loaded the dataset yet, you can do it like this:
# population = pd.read_csv("your_dataset.csv")

# Assuming you have a 'population' DataFrame with the mentioned columns
x = population["Population (2020)"]
y = population["Density (P/Km²)"]

# Perform linear regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

# Define the regression line function
def myfunc(x):
    return slope * x + intercept

# Calculate the predicted values
mymodel = list(map(myfunc, x))

# Plot the scatter plot and regression line
plt.scatter(x, y)
plt.plot(x, mymodel, color='red')  # Adding a regression line
plt.xlabel("Population (2020)")
plt.ylabel("Density (P/Km²)")
plt.title("Linear Regression: Population vs Density")
plt.show()
```
